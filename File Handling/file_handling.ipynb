```python
##write a file without overwriting

with open('example.txt','a') as file:
    file.write("Hello, How are you?\n")
    file.write("I am fine, how about you?\n")
    file.write("I am also fine, how about you?\n")
    file.write("I am super fine, how about you?")
```

```python
#Read a text file and count the number of lines, words, and characters.
# Counting lines, words, and characters in a text file

def count_number_of_words(file_path):
    with open(file_path, 'r') as file:
        lines=file.readlines()
        line_count=len(lines)
        word_count=sum(len(line.split()) for line in lines)
        char_count=sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path='example.txt'
line_count, word_count, char_count=count_number_of_words(file_path)
print(f"Number of lines: {line_count}")
print(f"Number of words: {word_count}")
print(f"Number of characters: {char_count}")
```

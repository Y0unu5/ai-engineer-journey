```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('diwali_sales.csv', encoding='unicode_escape')
df.head()
```

```python
df.info()
```

```python
##drop unrelated nlank cl
df.drop(['Status','unnamed1'], axis=1, inplace=True)

df.head()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
##change dt
df['Amount']=df['Amount'].astype(int)
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'User_ID':'Customer_ID'}, inplace=True)
df.rename(columns={'Age Group':'Age_Group'}, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
# use describe() for specific columns
df[['Age','Orders','Amount']].describe()
```

```python

gender_count=sns.countplot(x='Gender', data=df)

for x in gender_count.containers:
    gender_count.bar_label(x)
    

```

```python
# plotting a bar chart for gender vs total amount
sales_gender=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gender)
```

```python
ax=sns.countplot(x='Age_Group', data=df,hue='Gender')

for x in ax.containers:
    ax.bar_label(x)

    
```

```python
sales_by_age=df.groupby(['Age_Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age_Group', y='Amount', data=sales_by_age, color='black')
```

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states

total_sales_by_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.pie(total_sales_by_state['Amount'], labels=total_sales_by_state['State'], autopct='%1.2f%%')
plt.show()
```

```python
ax=sns.countplot(x='Marital_Status', data=df,hue='Gender')

for x in ax.containers:
    ax.bar_label(x)
```

```python
sales_state=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_by_occupation=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=sales_by_occupation, x='Occupation', y='Amount')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for x in ax.containers:
    ax.bar_label(x)
```

```python
sales_by_product=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=sales_by_product, x='Product_Category', y='Amount')
```

```python
df.head()
sales_by_pid=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_by_pid, x='Product_ID', y='Orders')
```

### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.
2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python

array1=np.random.randint(1,100,(5,5))
print("Random array",array1)


##mean
## mean=sum(elements in matrix)/total number of elements
print("Mean", np.mean(array1))


##median
## median=middle element if its odd then (n+1)/2
print("Median", np.median(array1))

##standard deviation
##std=variance**0.5
print("Standard deviation", np.std(array1))

##variance
##(element-mean)^2/total number of elements
print("Variance", np.var(array1))
```

```python
array=np.arange(1,10).reshape(3,3)
print("Original array",array)

mean=np.mean(array)
std=np.std(array)
normaloize=(array-mean)/std
print("Normalized array +++++++++++++++",normaloize)
```

### Assignment 10: Masked Arrays

1. Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.
2. Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.

```python
##1. Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.
arr=np.random.randint(1,16,(4,4))
print("Original array",arr)

mask=arr>10
print("Masked array",mask)

unmasked_arr=arr[mask]
print("Unmasked array",unmasked_arr)

sum=unmasked_arr.sum()
print("Sum of unmasked elements",sum)

##2. Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.
arr=np.random.randint(1,16,(3,3))
print("Original array",arr)

mask=np.eye(3,dtype=bool)
print("Masked array",mask)

unmasked_arr=arr[mask]
print("Unmasked array",unmasked_arr)

mean=unmasked_arr.mean()
print("Mean of unmasked elements",mean)

arr[mask]=mean
print("Modified array",arr)
```

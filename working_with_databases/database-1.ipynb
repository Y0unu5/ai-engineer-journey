```python
import sqlite3
```

```python
connection = sqlite3.connect("employees.db")

cursor=connection.cursor()
cursor.execute("DROP TABLE IF EXISTS employees")
cursor.execute("CREATE TABLE employees(first_name TEXT, last_name TEXT, department TEXT)")
connection.commit()



```

```python
##Insert some data
cursor.execute("INSERT INTO employees VALUES('John', 'Doe', 'Developer')")
cursor.execute("INSERT INTO employees VALUES('Jane', 'Doe', 'Banker')")
cursor.execute("INSERT INTO employees VALUES('John', 'Smith', 'Finance')")

connection.commit()
```

```python
##update data

cursor.execute('UPDATE employees SET department="IT Specialist" WHERE first_name="John" AND last_name="Doe"')
connection.commit()
```

```python
##delete

cursor.execute('DELETE FROM employees WHERE first_name="John" AND last_name="Doe"')
connection.commit()
```

📌 Project Overview
This report provides a comprehensive analysis of coffee sales from a vending machine. It explores revenue trends, customer behaviors, and key sales patterns to support data-driven business decisions.

📝 Executive Summary
This report analyzes sales data from a coffee vending machine to uncover customer behavior, sales trends, and product performance.

Key findings include:

Most sales occur in the morning hours (9 AM to 11 AM), suggesting high demand during typical coffee break times.
The vending machine generates a steady revenue stream, with clear patterns in best-selling coffee products.
Payment method analysis shows a preference for card payment.
Loyal customers make repeat purchases, opening up opportunities for customer retention strategies.
Sales fluctuate throughout the week and month, providing insights to optimize promotions and stock levels.
Overall, the data-driven insights suggest actionable strategies to boost sales, improve customer experience, and streamline vending machine operations.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("coffee_sales.csv")
df.head()
```

```python
##check data types
df.info()
```

```python
##removing datetime
df['date']=pd.to_datetime(df['date'])

df['time']=pd.to_datetime(df['datetime']).dt.strftime('%H:%M:%S')
df.drop('datetime',axis=1,inplace=True) ##drop old columns


df.head()
```

```python
# Total Revenue and average spending per transaction

total_revenue=df['money'].sum()
avg_spending=df['money'].mean()
print(f"Total Revenue: ${total_revenue:.2f}")
print(f"Average Spending Per Transaction: ${avg_spending:.2f}")
```

```python
# Daily sales chart

daily_sales=df.groupby('date')['money'].sum()
plt.figure(figsize=(12,6))
plt.plot(daily_sales.index,daily_sales.values, color='blue', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Daily Sales')
plt.title('Daily Sales Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Best-selling product analysis

df['coffee_name'].value_counts()
```

```python
# Best Selling Coffee Chart

plt.figure(figsize=(10,5))
sns.barplot(x=df['coffee_name'].value_counts().index, y=df['coffee_name'].value_counts().values, palette='Set2')
plt.xlabel('Coffee Name')
plt.ylabel('Number of Sales')
plt.title('Best Selling Coffee')
plt.tight_layout()
plt.show()
```

```python
# Total revenue for each type of coffee
df.groupby('coffee_name').agg({'money': 'sum'}).sort_values(by='money', ascending=False)
##df.groupby('coffee_name')['money'].sum().sort_values(ascending=False)
```

```python
# Revenue chart for each type of coffee.
coffee_revenue = df.groupby("coffee_name")["money"].sum().sort_values(ascending=False)
plt.figure(figsize=(13,6))
sns.barplot(x=coffee_revenue.index, y=coffee_revenue.values, palette='viridis')
plt.title("Revenue per type of coffee")
plt.tight_layout()
plt.show()
```

```python
# Payment methods distribution account
df['cash_type'].value_counts()
```

```python
# pie chart of payment methods
plt.figure(figsize=(6,6))
plt.pie(df['cash_type'].value_counts(), labels=df['cash_type'].value_counts().index, autopct='%1.1f%%', startangle=30, pctdistance=0.85)
plt.axis('equal')
plt.title('Distribution of payment methods')
plt.show()
```

```python
# Take the top 10 customers by number of purchases
df['card'].value_counts().head(10)
```

```python
# Chart of repeat customers

top_customers=df['card'].value_counts().head(20)
plt.figure(figsize=(20,10))
sns.barplot(x=top_customers.index, y=top_customers.values, palette='rocket')
plt.xticks(rotation=90)
plt.title("Most frequent customers")
plt.show()
```

```python
df.head()
# Extract hours from `datetime`
df['hour'] = pd.to_datetime(df['datetime']).dt.hour

# the number of sales per hour
sales_by_hour = df.groupby("hour").size()
print(sales_by_hour)
```

```python
# Peak time chart

plt.figure(figsize=(12,6))
df.groupby('hour').size().plot(marker='o',color='r')
plt.xticks(range(0,24))
plt.ylabel('Number of Sales')
plt.title('Peak Sales Time')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
##extract weekday
df['weekday']=pd.to_datetime(df['date']).dt.day_name()
##total revenue on that day
df.groupby('weekday')['money'].sum()
```

```python
# Arrange the days correctly
sales_weekday=df.groupby('weekday')['money'].sum()

date_order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sales_weekday=sales_weekday.reindex(date_order)

sales_weekday
```

```python
##sales chart by day
plt.figure(figsize=(12,6))
sns.barplot(x=sales_weekday.index, y=sales_weekday.values, palette='Set2')
plt.xlabel('Weekday')
plt.ylabel('Daily Sales')
plt.title('Sales by Day')
plt.tight_layout()
plt.show()
```

```python
df['month']=pd.to_datetime(df['date']).dt.month
df.groupby('month')['money'].sum()
```

```python
monthly_sales=df.groupby('month')['money'].sum()

plt.figure(figsize=(12,6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette='Set2')
plt.xlabel('Month')
plt.ylabel('Monthly Sales')
plt.title('Sales by Month')
plt.tight_layout()
plt.show()
```

```python
# Classify transactions into morning (before 12 o'clock) or evening
df['time_of_day'] = df['hour'].apply(lambda x: 'morning' if x < 12 else 'evening')

# Group by time of day and calculate total sales    
sales_by_time = df.groupby('time_of_day')['money'].sum()

sales_by_time
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=sales_by_time.index, y=sales_by_time.values, palette='Set2')
plt.xlabel('Time of Day')
plt.ylabel('Total Sales')
plt.title('Sales by Time of Day')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Consider offering promotions during non-peak hours to balance the sales load.


offer_for_sales=df['time_of_day'].apply(lambda x: '10 percent offer' if x=='morning' else '5 percent offer')
df['offer_for_sales']=offer_for_sales

##apply 10% for morning sales
ten_per_morning=df['offer_price_morning']=df['money']- df['money']*0.1
##apply 5% for evening sales
five_per_evening=df['offer_price_evening']=df['money']- df['money']*0.05

ten_per_morning
# five_per_evening

df.head()
```

```python

df['card'].value_counts()
# Analyze customer cards further to identify demographics behind loyal customers

loyal_customers=df['card'].value_counts().head(10)

df.groupby('card')['money'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=loyal_customers.index, y=loyal_customers.values, palette='Set2')
plt.xlabel('Customer Card')
plt.ylabel('Number of Sales')
plt.title('Loyal Customers')
plt.tight_layout()
plt.show()
```

```python
# Monitor the least popular products to assess if they should be replaced or promoted differently
count=df.groupby('coffee_name')['money'].sum().sort_values(ascending=True).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=count.index, y=count.values,
    palette='Set2')
plt.xlabel('Coffee Name')
plt.ylabel('Sales')
plt.title('Least Popular Coffee')
plt.tight_layout()
plt.show()
```

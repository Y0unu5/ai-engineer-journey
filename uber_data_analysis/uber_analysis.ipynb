```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
import datetime
```

```python
df= pd.read_csv('uber_data.csv')
df.head()
```

```python
df.info()
```

```python
##if there any data points missing
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df['START_DATE*']=pd.to_datetime(df['START_DATE*'],format="%m/%d/%Y %H:%M",errors='coerce')
df['END_DATE*']= pd.to_datetime(df['END_DATE*'],format="%m/%d/%Y %H:%M",errors='coerce')
```

```python
hour=[]
day=[]
dayofweek=[]
month=[]
weekday=[]


df['HOUR'] = df['START_DATE*'].dt.hour
df['DAY'] = df['START_DATE*'].dt.day
df['DAYOFWEEK'] = df['START_DATE*'].dt.dayofweek
df['MONTH'] = df['START_DATE*'].dt.month

df['WEEKDAY'] = df['START_DATE*'].dt.day_name()

df.head()
```

```python
df['CATEGORY*'].value_counts()
```

```python
sns.countplot(x='CATEGORY*',data=df)
```

How long do people travel with Uber?

```python
df['MILES*'].plot.hist()
```

What Hour Do Most People Take Uber To Their Destination?

```python
hours=df['START_DATE*'].dt.hour.value_counts()
hours.plot(kind='bar',color='red', figsize=(10,5))
plt.xlabel('HOUR')
plt.ylabel('Frequency')
plt.title('Number of rides per hour')
plt.show()
```

Check The Purpose Of Trips

```python
df['PURPOSE*'].value_counts().plot(kind='bar',color='brown', figsize=(10,5))
```

Which Day Has The Highest Number Of Trips

```python
df['WEEKDAY'].value_counts().plot(kind='bar',color='green', figsize=(10,5))
```

What Are The Number Of Trips Per Each Day?

What Are The Trips In The Month
